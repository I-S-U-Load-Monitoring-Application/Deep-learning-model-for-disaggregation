```python
import pandas as pd
import os
import glob

DATA_DIR = 'dataset/REFIT/CLEAN_REFIT/'
appliance_name = 'Kettle'  # Change as needed
appliance_threshold = 10   # Power threshold to consider appliance ON

aggregate_readings = []

# Loop through all house CSV files
for filepath in glob.glob(os.path.join(DATA_DIR, 'House_*.csv')):
    df = pd.read_csv(filepath, parse_dates=['timestamp'], index_col='timestamp')
    
    if appliance_name not in df.columns:
        continue  # Appliance not in this house
    
    # Filter times when appliance is ON
    on_df = df[df[appliance_name] > appliance_threshold]
    
    if not on_df.empty:
        aggregate_readings.append(on_df['Aggregate'])

# Combine all aggregate readings
all_agg = pd.concat(aggregate_readings)

# Compute stats
agg_mean = all_agg.mean()
agg_std = all_agg.std()

print(f"{appliance_name} AGGREGATE MEAN: {agg_mean:.2f}")
print(f"{appliance_name} AGGREGATE STD: {agg_std:.2f}")
```

```python

import os
import pandas as pd
from glob import glob

# Set your dataset directory
data_dir = "dataset/REFIT/CLEAN_REFIT/"
house_files = glob(os.path.join(data_dir, "*.csv"))

# Dictionary to store readings per appliance across all houses
appliance_data = {f'Appliance{i}': [] for i in range(1, 10)}

# Step 1–3: Aggregate data
for file in house_files:
    df = pd.read_csv(file)
    for appliance in appliance_data:
        if appliance in df.columns:
            appliance_data[appliance].extend(df[appliance][df[appliance] > 0].values)

# Step 4: Compute thresholds using the 5th percentile
appliance_thresholds = {}
for appliance, values in appliance_data.items():
    if len(values) > 0:
        series = pd.Series(values)
        threshold = series.quantile(0.05)  # Or try 0.03 or 0.10 depending on the appliance
        appliance_thresholds[appliance] = round(threshold, 2)
    else:
        appliance_thresholds[appliance] = None

# Print results
for app, thresh in appliance_thresholds.items():
    print(f"{app}: {thresh} W")
```

```python
#MAXIMUM POWER ON
import pandas as pd

appliance_name = 'fridge'
data_path = 'dataset/REFIT/CLEAN_REFIT/'
house_ids = [2, 5, 9, 12, 15]
channel_mapping = {2:1, 5:1, 9:1, 12:1, 15:1}
on_threshold = 40  # Only consider values above this as "on"

max_values = {}
mean_values = {}
std_values = {}

for house_id in house_ids:
    file_path = f'{data_path}CLEAN_House{house_id}.csv'
    channel = channel_mapping[house_id]
    try:
        df = pd.read_csv(
            file_path,
            usecols=[2, channel + 2],
            names=['aggregate', appliance_name],
            header=0
        )
        on_values = df[df[appliance_name] > on_threshold][appliance_name]
        max_power = on_values.max()
        mean_power = on_values.mean()
        std_power = on_values.std()
        max_values[f'House {house_id}'] = max_power
        mean_values[f'House {house_id}'] = mean_power
        std_values[f'House {house_id}'] = std_power
        
    except Exception as e:
        print(f"Failed to load House {house_id}: {e}")

# Display the max values
for house, power in max_values.items():
    print(f"{house} - Max ON Power: {power:.2f} W")
```

```python
import pandas as pd
import os

def detect_tv_channel(file_path, appliance_name="television", power_threshold=(50, 500), max_power_cutoff=4000):
    print(f"\nAnalyzing file: {file_path}")
    
    df = pd.read_csv(file_path)
    df = df.dropna(axis=1, how='all')  # Drop any empty columns

    # Skip first two columns: time and aggregate
    candidates = []
    for i, col in enumerate(df.columns[2:], start=2):
        power_data = df.iloc[:, i]
        mean_power = power_data.mean()
        std_power = power_data.std()
        max_power = power_data.max()

        print(f"Channel {i-2} (Column: {col}) → Mean: {mean_power:.2f} W, Max: {max_power:.2f} W, Std: {std_power:.2f} W")

        if power_threshold[0] <= mean_power <= power_threshold[1] and max_power <= max_power_cutoff:
            candidates.append((i-2, mean_power, std_power, max_power))

    if candidates:
        print("\nLikely TV channels based on mean/max power thresholds:")
        for c in candidates:
            print(f"  → Channel {c[0]} | Mean: {c[1]:.2f} W | Max: {c[3]:.2f} W")
    else:
        print("⚠️ No channels matched the expected TV power profile.")
    
    return candidates

# Example usage
# Adjust the file path as needed for other houses
file_path = "dataset/REFIT/CLEAN_REFIT/CLEAN_House6.csv"
tv_candidates = detect_tv_channel(file_path)
```

```python
import os
import pandas as pd
from typing import Dict

class SimpleNILMDataSet:
    def __init__(self, root_dir: str):
        self.root_dir = root_dir

    def load_appliance_series(self, house: int, channel: int, nrows=None) -> pd.Series:
        """
        Load the power series for a given house and channel.
        Assumes files are named as house_{n}/channel_{m}.csv
        """
        filepath = os.path.join(self.root_dir, f'CLEAN_', f'house{house}.csv')
        if not os.path.exists(filepath):
            raise FileNotFoundError(f"No file found at: {filepath}")

        # Try common formats
        try:
            df = pd.read_csv(filepath, nrows=nrows, names=['Time', 'power'], header=None)
        except Exception as e:
            raise ValueError(f"Failed to read {filepath}: {e}")

        # Convert timestamp to datetime if it's in seconds
        df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
        df.set_index('timestamp', inplace=True)

        return df['power']
```
